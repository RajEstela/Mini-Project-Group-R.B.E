## Merging Both Datasets

We will merge the clean_match_data with clean_champions_data based on the name of the champion, and then adding the class of that champion for each champion.

````python
# dataset 1 - clean version
display(clean_match_data)
````

````python
# dataset 2 - clean version
display(clean_champion_data)
````

````python
# merging both datasets using for loop
for i in range(1, 3):
    team = "t" + str(i)

    for j in range(1, 6):
        position = "p" + str(j)
        champion = team + position + "_champion"
        champion_class = team + position + "_class"

        if i == 1 and j == 1:
            merged_data = clean_match_data

        merged_data = pd.merge(merged_data, clean_champion_data, left_on=champion, right_on='Name', how='left')
        merged_data.rename(columns={'Class': champion_class}, inplace=True)
        merged_data.drop('Name', axis=1, inplace=True)

display(merged_data)
````

The merged dataset is correct, but we will reorder the columns so that the respective columns will be after the respective champion for better clarity.

````python
columns_order = ['t1_result', 't2_result', 't1p1_champion', 't1p1_class', 't1p2_champion', 't1p2_class', 't1p3_champion', 't1p3_class', 't1p4_champion', 't1p4_class', 't1p5_champion', 't1p5_class', 't2p1_champion', 't2p1_class', 't2p2_champion', 't2p2_class', 't2p3_champion', 't2p3_class', 't2p4_champion', 't2p4_class', 't2p5_champion', 't2p5_class', 't1_totalgold', 't2_totalgold', 't1_barons', 't2_barons', 't1_dragons', 't2_dragons', 't1_heralds', 't2_heralds', 't1_towers', 't2_towers', 't1_inhibitors', 't2_inhibitors', 't1_kills', 't2_kills', 't1_totalcs', 't2_totalcs']

merged_data = merged_data[columns_order]

display(merged_data)
````

The merged dataset is in the correct order and we will create a new csv file based on that dataset (and for further processing if needed) in the EDA section.

````python
merged_data.to_csv("datasets/cleaned_dataset.csv", index=False)
````
